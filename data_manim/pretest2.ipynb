```python
# Question4
# bmi컬럼의 각 결측치들을 결측치를 가진 환자 나이대(10단위)의 평균 bmi 값으로 대체한 후 대체된 bmi 컬럼의 평균을 소숫점 이하 3자리 까지 구하여라
```

```python
df["age"] = df["age"].str.replace('*','').astype(int)
df["aged"] = (df["age"]/10).astype(int)
```

```python
val = df.groupby("aged")["bmi"].mean()
df.loc[df["bmi"].isnull(), ["bmi"]]=val[df.loc[df["bmi"].isnull(), "aged"].values].values
```

```python
# Question10
# 각 세대 중(Generation 컬럼)의 Speed와 Defense 컬럼의 차이(절댓값)이 가장 큰 세대는?
```

```python
def custom(df):
    return max(abs(df["Speed"]-df["Defense"]))

df.groupby("Generation").apply(custom).sort_values().tail(1).index[0]
```
